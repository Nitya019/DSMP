```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100  # number of trials, eg : tossing a coin 10 times
p = 0.5
size = 500 #number of samples, eg: we will perform the experiment 100 times

binomial_dist = np.random.binomial(n,p,size)

plt.hist(binomial_dist, density= True)

plt.show()
```

```python
# for a biased coin

n = 100  # number of trials, eg : tossing a coin 10 times
p = 0.9
size = 500 #number of samples, eg: we will perform the experiment 100 times

binomial_dist = np.random.binomial(n,p,size)

plt.hist(binomial_dist, density= True)

plt.show()
```

*The graph has shifted towards the right*

```python
n = 100  # number of trials, eg : tossing a coin 10 times
p = 0.1
size = 500 #number of samples, eg: we will perform the experiment 100 times

binomial_dist = np.random.binomial(n,p,size)

plt.hist(binomial_dist, density= True)

plt.show()
```
