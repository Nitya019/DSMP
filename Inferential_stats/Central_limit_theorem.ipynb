```python
import numpy as np
import matplotlib.pyplot as plt
```

#### *CLT on population following Uniform distribution*

```python
num_sample = 1000
sample_size = 30
dist_range = (0,1)

#generating samples from a population that follows uniform distribtuion

samples = np.random.uniform(dist_range[0], dist_range[1], (num_sample, sample_size))
samples.shape
```

```python
#calculating sample means and finding the sample distribution of sample means

sample_mean = np.mean(samples, axis = 1)

plt.hist(sample_mean, bins = 20, density= True)
plt.title("Sample distribution of sample mean")
plt.xlabel("Sample mean")
plt.ylabel("Density")
plt.show()
```

*Close to a normal distribution*

#### *CLT on population following Exponential distribution*

```python
num_sample = 1000
sample_size = 30
lambda_p = 2
#generating samples from a population that follows exponential distribtuion

samples = np.random.exponential(scale = 1/lambda_p, size =  (num_sample, sample_size))
samples.shape
```

```python
#calculating sample means and finding the sample distribution of sample means

sample_mean = np.mean(samples, axis = 1)

plt.hist(sample_mean, bins = 20, density= True)
plt.title("Sample distribution of sample mean")
plt.xlabel("Sample mean")
plt.ylabel("Density")
plt.show()
```

#### Prove : Mean of sample distribution is equal to that of population distribution

```python
import numpy as np

# Set the parameters
num_samples = 10000
sample_size = 50

# Gamma distribution parameters
gamma_shape = 2
gamma_scale = 1

# Calculate the theoretical mean and variance
theoretical_mean = gamma_shape * gamma_scale
theoretical_variance = (gamma_shape * gamma_scale ** 2)

# Generate samples from the Gamma distribution
samples = np.random.gamma(shape=gamma_shape, scale=gamma_scale, size=(num_samples, sample_size))

# Calculate the sample means
sample_means = np.mean(samples, axis=1)

# Calculate the empirical mean and variance of the sample means
empirical_mean = np.mean(sample_means)
empirical_variance = np.var(sample_means)

# Compare the theoretical and empirical values
print(f"Theoretical mean: {theoretical_mean:.4f}")
print(f"Empirical mean: {empirical_mean:.4f}")
print(f"\n")
print(f"Theoretical variance: {theoretical_variance:.4f}")
print(f"Empirical variance: {empirical_variance:.4f}")
```
